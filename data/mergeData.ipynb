```python
import numpy as np
import pandas as pd
```

```python
train_drcd = pd.read_csv('train_drcd.csv')
dev_drcd = pd.read_csv('dev_drcd.csv')
train_drcd['答案'] += 1
dev_drcd['答案'] += 1
train_drcd_values = train_drcd.values
dev_drcd_values = dev_drcd.values
print(train_drcd_values)
```

```python
train_our_values = pd.read_csv('train.csv').values
dev_our_values = pd.read_csv('dev.csv').values
```

```python
all = np.concatenate((train_drcd_values, dev_drcd_values, train_our_values, dev_our_values), axis=0)
all[:,0] = np.arange(all.shape[0])
print(all)
np.random.shuffle(all)
print(all)
```

```python
valid = int(all.shape[0] * 0.1)
validation = all[:valid,:]
train = all[valid:, :]
print(validation.shape)
print(train.shape)
```

```python
train_df = pd.DataFrame(train, columns=['id', 'article', 'question', 'op0', 'op1', 'op2', 'op3', 'label'])
train_df.to_csv('train_merge.csv', index=False, sep=',')
dev_df = pd.DataFrame(validation, columns=['id', 'article', 'question', 'op0', 'op1', 'op2', 'op3', 'label'])
dev_df.to_csv('dev_merge.csv', index=False, sep=',')
```
